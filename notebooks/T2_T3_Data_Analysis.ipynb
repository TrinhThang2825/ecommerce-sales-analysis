```python
import pandas as pd
import numpy as np

# Thiết lập số lượng dòng dữ liệu giả lập
n_rows = 200
np.random.seed(42)

# Tạo dữ liệu giả lập
data = {
    'Laptop_ID': [f'LP{i:03d}' for i in range(1, n_rows + 1)],
    'Brand': np.random.choice(['Apple', 'Dell', 'HP', 'Asus', 'Lenovo'], n_rows),
    'Type': np.random.choice(['Workstation', 'Gaming', 'Ultrabook', 'Office'], n_rows),
    'RAM_GB': np.random.choice([8, 16, 32, 64], n_rows),
    'Storage_GB': np.random.choice([256, 512, 1024], n_rows),
    'Rating': np.round(np.random.uniform(3.5, 5.0), 1)
}

df = pd.DataFrame(data)

# Logic tạo giá: RAM càng cao giá càng đắt
df['Price_VND'] = (df['RAM_GB'] * 700000) + (df['Storage_GB'] * 12000) + np.random.randint(6000000, 10000000, n_rows)
df.loc[df['Brand'] == 'Apple', 'Price_VND'] += 9000000

# Xuất file CSV
df.to_csv('laptop_data_2026.csv', index=False)
print("✅ Đã tạo file laptop_data_2026.csv thành công!")
```
